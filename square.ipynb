```python
import scipy.signal as sig
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [20, 10]
import numpy as np
```

```python
t = np.linspace(0, 42e-6, int(2e9*14e-6*3))
a = sig.square(2 * np.pi * 1/14e-6 * t, duty=0.02857)
a2 = sig.square(2 * np.pi *  1/14e-6 * t, duty=0.02857)
plt.plot(a, label="m=1")
plt.plot(a2, label="m=2")
plt.legend()
plt.show()
```
