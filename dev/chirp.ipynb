```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t_tx = 60e-6
fc = 10e9
fs = 600e6
bw = 300e6
oversample = 10
tau = 9e-6
t = np.linspace(0, t_tx, int(t_tx * fs * oversample))
n = np.arange(0, 2500)
lamb = 300e6 / fc
```

```python
sig = np.exp(1j* np.pi * bw/t_tx * t**2)
plt.plot(t*1e6, sig)
plt.xlabel("Time ($\mu$s)")
plt.ylabel("Amplitude")
plt.title("Chirp amplitude over time")
#plt.savefig("chirp_tx_sig.pdf")
```

```python
freq = bw / t_tx * t
plt.plot(t*1e6, freq/1e6)
plt.xlabel("Time ($\mu$s)")
plt.ylabel("Frequency (MHz)")
plt.title("Chirp frequency over time")
plt.grid()
plt.savefig("inst_up_chirp_tx.pdf")
```
