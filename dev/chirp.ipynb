```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t_tx = 1e-6
fs = 600e6
bw = 300e6
oversample = 10
t = np.linspace(15e-6, 16e-6, int(t_tx * fs * oversample))
```

```python
sig = np.cos(2*np.pi + np.pi * bw/t_tx * t**2)
plt.plot(t*1e6, sig)
plt.xlabel("Time ($\mu$s)")
plt.ylabel("Amplitude")
plt.title("Chirp amplitude over time")
plt.show()
```

```python
freq = bw / t_tx * t
plt.plot(t*1e6, freq/1e8)
plt.xlabel("Time ($\mu$s)")
plt.ylabel("Frequency (MHz)")
plt.title("Chirp frequency over time")
plt.show()
```
